# Simple Linear Regression

#### Importing the libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

#### Importing the dataset and cleaning

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

```python
insurance_data.isnull().sum() # Checking for NaN values in our dataset
```

```python
insurance_data.shape # getting the size of our dataset
```

```python
# checking for duplicate rows
dup_rows = insurance_data[insurance_data.duplicated()]
print(dup_rows)
```

```python
# drop the duplicated row
insurance_data = insurance_data.drop_duplicates()
insurance_data.shape
```

```python
# Showing statistic analysis with 'describe()' function
insurance_data.describe()
```

#### Selecting rows and columns for our dependent and independent variables

```python
x = insurance_data.iloc[:, 0].values
y = insurance_data.iloc[:, -1].values
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

#### Showing a scatter plot of our data

```python
plt.scatter(x, y, color='blue')
plt.title('How Age Affect Insurance Cost')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

#### Creating a Simple Linear Regression Model and plotting the best fit line

```python
charge_model = LinearRegression()
charge_model.fit(x,y)


y_pred = charge_model.predict(x)
plt.scatter(x, y, color='blue')
plt.plot(x, charge_model.predict(x), color='red')
plt.title('How Age Affect Insurance Cost')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
